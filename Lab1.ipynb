```python
#replaced entries in 'Gender' to have a simplified version of the categories 

def clean(x):
    if x in ['Male', 'MALE']:
        return 'M'
    elif x in ['F','female','Femal']:
        return 'F'
    else:
        return 'U'
data['gender'] = list(map(clean, data['gender']))
```

```python
#replaced entries in 'st' to have a simplified version of the categories 

def clean(x):    
    if x in ['Washington', 'WA']:
        return 'WA'
    elif x in ['Arizona','AZ']:
        return 'AZ'
    elif x in ['California','Cali']:
        return 'CA'
    elif x in ['Nevada']:
        return 'NV'
    elif x in ['Oregon']:
        return 'OR'
    else:
        return 'U'
data['st'] = list(map(clean, data['st']))
```
